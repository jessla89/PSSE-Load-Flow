##### Load Flow Analysis For 'Load-MW = 3264, Gen-MW = 3324.8, Reactors-Mvar = 600, Capacitors-Mvar = 950, Load-Mvar = 1950'

```python
import pandas as pd
import numpy as np
```

```python
f = open("new_2.csv", "r")
data_new = f.read()
```

```python
header_string = data_new.split('\n')[0]
```

```python
header = header_string.split(';')
```

```python
info_string_list = data_new.split('\n')[1:]
```

```python
info_list = []
```

```python
for each_string in info_string_list:
    info_list.append(each_string.split(';'))
```

```python
df = pd.DataFrame(info_list)
df.columns = header 
```

```python
df
```

```python
df = df.drop(index = (len(df) -1))
```

```python
df
```

```python
df['Bus Name'] = df['Bus Name'].replace(r'^\s*$', np.nan, regex=True).ffill()
```
